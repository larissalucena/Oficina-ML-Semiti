# Import das bibliotecas pandas e numpy

# Descarte de colunas

```python
dados = dados.drop(['ano','nome','sequencial_candidato','cargo', 'grau','ocupacao','estado_civil'], axis=1)
```

# Separação entre features (X) e variável-resposta (y) 

```python
y = dados.situacao
X = dados.drop(['situacao'],axis=1)
```

```python
X.head()
```

# Dummyfying - one-hot encoding

```python
X = pd.get_dummies(X)
```

```python
X.head()
```

# Label encoding de y

```python
y.head()
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
classes = ['nao_eleito','eleito']
le.fit(classes)
y = le.fit_transform(y)
```

```python
print(y)
```

# Divisão entre conjuntos de treino e teste

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state = 42)
```

```python
print(len(X_train))
print(len(X_test))
print(len(y_train))
print(len(y_test))
```

# Classificações

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
log = LogisticRegression()
```

```python
log.fit(X_train,y_train)
```

```python
y_pred = log.predict(X_test)
```

```python
import sklearn.metrics as met
```

```python
test_accuracy = met.accuracy_score(y_test, y_pred)
test_precision = met.precision_score(y_test, y_pred)
test_recall = met.recall_score(y_test, y_pred)
test_fmeasure = met.f1_score(y_test, y_pred)
test_classification_report = met.classification_report(y_test, y_pred)
test_conf_matrix = met.confusion_matrix(y_test, y_pred)

print("Resultados - Logistic Regression")
print("Acuracy: ", test_accuracy)
print("Precision: ", test_precision)
print("Recall: ", test_recall)
print("F-Measure: ", test_fmeasure)
print('Confusion Matrix: ')
print(test_conf_matrix)
print('Classification report: ')
print(test_classification_report)
```

# SVM

```python
from sklearn.svm import SVC
svm = SVC()
svm.fit(X_train,y_train)
y_pred = svm.predict(X_test)
```

```python
test_accuracy = met.accuracy_score(y_test, y_pred)
test_precision = met.precision_score(y_test, y_pred)
test_recall = met.recall_score(y_test, y_pred)
test_fmeasure = met.f1_score(y_test, y_pred)
test_classification_report = met.classification_report(y_test, y_pred)
test_conf_matrix = met.confusion_matrix(y_test, y_pred)

print("Resultados - SVM")
print("Acuracy: ", test_accuracy)
print("Precision: ", test_precision)
print("Recall: ", test_recall)
print("F-Measure: ", test_fmeasure)
print('Confusion Matrix: ')
print(test_conf_matrix)
print('Classification report: ')
print(test_classification_report)
```

# Random Forest Classifier 

```python
from sklearn.ensemble import RandomForestClassifier 

rf = RandomForestClassifier()
rf.fit(X_train, y_train)

y_pred = rf.predict(X_test)

 
test_accuracy = met.accuracy_score(y_test, y_pred)
test_precision = met.precision_score(y_test, y_pred)
test_recall = met.recall_score(y_test, y_pred)
test_fmeasure = met.f1_score(y_test, y_pred)
test_classification_report = met.classification_report(y_test, y_pred)
test_conf_matrix = met.confusion_matrix(y_test, y_pred)

print("Resultados - Random Forest")
print("Acuracy: ", test_accuracy)
print("Precision: ", test_precision)
print("Recall: ", test_recall)
print("F-Measure: ", test_fmeasure)
print('Confusion Matrix: ')
print(test_conf_matrix)
print('Classification report: ')
print(test_classification_report)
```
